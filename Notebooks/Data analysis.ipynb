## Summary

The dataset contains information about **4,809 movies** with **24 features** related to movie details, financial performance, ratings, and production information. The attributes include movie budget, revenue, genres, popularity, cast, crew, release date, runtime, vote average, and vote count. It also contains textual information such as overviews, taglines, keywords, and production details. Most movies in the dataset are in English, and the dataset combines both numerical and categorical data, making it suitable for exploratory data analysis, recommendation systems, sentiment analysis, and movie success prediction.


## Column Description

* `genres`: The category or categories the film falls into (e.g., Action, Science Fiction)
* `homepage`: The URL to the official website of the movie.
* `id`: A unique numerical identifier for the movie (usually from The Movie Database - TMDB).
* `keywords`: Specific words or short phrases that describe the movie's themes, plot, or setting.
* `original_language`: The primary language the film was originally shot and released in (e.g., 'en' for English).
* `original_title`: The native title of the movie before any regional translations.
* `overview`: A brief text summary or synopsis of the movie's plot.
* `popularity`: A metric scoring the movie's cultural footprint and web traffic based on audience engagement.
* `production_companies`: The production studios or companies that funded and created the movie.
* `production_countries`: The countries where the movie was officially produced.
* `release_date`: The date the movie was first released to the public.
* `revenue`: The total worldwide box office earnings generated by the movie (in USD).
* `runtime`: The total length or duration of the movie in minutes.
* `spoken_languages`: The specific languages that are spoken by characters within the film.
* `status`: The current release status of the film (e.g., "Released", "Post Production", or "Rumored").
* `tagline`: The promotional catchphrase or marketing slogan used for the movie.
* `title`: The commonly used (often English) title of the movie.
* `vote_average`: The average rating given by viewers, typically on a scale from 0 to 10.
* `vote_count`: The total number of individual user votes or ratings the movie received.
* `movie_id`: A duplicate or secondary unique identifier (often used as a key to merge with other datasets like cast/crew).
* `cast`: A detailed list of the actors in the film and the specific characters they played.
* `crew`: A detailed list of the behind-the-scenes personnel (directors, writers, producers, etc.) and their specific jobs.

```python
movies_df = movies.copy()
```

```python
#dropping all the duplicate values
movies_df.dropna(inplace = True)
```

## genres

```python
movies_df['genres'].head(1)
```

```python
#function to extract the genre of the movie
def converter(row):
    l = []
    for i in ast.literal_eval(row):
        l.append(i['name'])
    return l
```

```python
movies_df['genres'] = movies_df['genres'].apply(converter)
```

```python
movies_df['genres'].head()
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(converter)
```

## cast

```python
#function to fetch top 3 cast members of the movie
def converter(row):
    l = []
    counter = 0
    for i in ast.literal_eval(row):
        if counter != 3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies_df['cast'] = movies_df['cast'].apply(converter)
```

```python
movies_df['cast'].head()
```

## Crew

```python
#function to extract the name of director 
def converter(row):
    l = []
    for i in ast.literal_eval(row):
        if i['job'] == 'Director':
            return i['name']
    return l
```

```python
movies_df['crew'] = movies_df['crew'].apply(converter)
```

```python
movies_df.info()
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df.head(5)
```

```python
#Function to contact the word in given columns
movies_df['cast'] = movies_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['genres'] = movies_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['crew'] = movies_df['crew'].str.replace(' ','').apply(lambda x: [x])
```

```python
movies_df.head(2)
```

```python
movies_df['tags'] = movies_df['overview']+ movies_df['keywords'] + movies_df['genres'] + movies_df['cast']+ movies_df['crew']
```

```python
movies_df.head(1)
```

```python
movies_df = movies_df[['id','tags','title']]
```

```python
movies_df['tags']=movies_df['tags'].str.join(" ")
```

```python
movies_df.head()
```

```python
movies_df['tags']= movies_df['tags'].apply(lambda x:x.lower() if  isinstance(x,str) else x)
```
